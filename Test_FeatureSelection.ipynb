```python
from sklearn import linear_model
from sklearn.metrics import accuracy_score
import numpy

def predict_ppdemand_li1(XC,field1,field2,field3,field4,field5):
    sdata = create_sumdata()
    lst_avg = [0,0,0,0,0]
    fields = [field1,field2,field3,field4,field5]
    nfields = []
    XC1 = []
    for i in range(5):
        if XC[i] != 0:
            nfields.append(fields[i])
            XC1.append(XC[i])

    print("Input Attributes:",nfields)
    print("Input Field:",XC1)
    X1 = numpy.array(sdata[nfields])
    #print(numpy.array(sdata[[field1,field2,field3,field4,field5]]))
    #liblinear
    Y1 = numpy.array(sdata['HP Types'])
    #regr1 = linear_model.LogisticRegression(solver='saga', random_state=0)
    regr1 = linear_model.LogisticRegression(solver='liblinear', random_state=0)
    regr1.fit(X1, Y1)
    y_pred = regr1.predict(numpy.array(sdata[nfields]))
    y_true = numpy.array(sdata['HP Types'])
    bnumber = regr1.predict_proba(numpy.array([XC1]))
    ascore = regr1.score(X1, Y1)
    ac_score = accuracy_score(y_true, y_pred)

    lst_avg[0] = round(ascore,3)
    lst_avg[1] = round(bnumber[0][0]*100,2)
    lst_avg[2] = round(bnumber[0][1]*100,2)
    lst_avg[3] = round(bnumber[0][2]*100,2)
    print("---Liblinear---")
    print("Actual Types:",y_true)
    print("PredictTypes:",y_pred)
    print("AccuracyScore:",ac_score)
    print("Test Score:",lst_avg[0])
    print("HP1:",lst_avg[1])
    print("HP2:",lst_avg[2])
    print("HP3:",lst_avg[3])


    return lst_avg

def predict_ppdemand_saga1(XC,field1,field2,field3,field4,field5):
    sdata = create_sumdata()
    lst_avg = [0,0,0,0,0]
    fields = [field1,field2,field3,field4,field5]
    nfields = []
    XC1 = []
    for i in range(5):
        if XC[i] != 0:
            nfields.append(fields[i])
            XC1.append(XC[i])

    print("Input Attributes:",nfields)
    print("Input Field:",XC1)
    X1 = numpy.array(sdata[nfields])
    #print(numpy.array(sdata[[field1,field2,field3,field4,field5]]))
    #liblinear
    Y1 = numpy.array(sdata['HP Types'])
    #regr1 = linear_model.LogisticRegression(solver='saga', random_state=0)
    regr1 = linear_model.LogisticRegression(solver='saga', random_state=0)
    regr1.fit(X1, Y1)
    y_pred = regr1.predict(numpy.array(sdata[nfields]))
    y_true = numpy.array(sdata['HP Types'])
    bnumber = regr1.predict_proba(numpy.array([XC1]))
    ascore = regr1.score(X1, Y1)
    ac_score = accuracy_score(y_true, y_pred)

    lst_avg[0] = round(ascore,3)
    lst_avg[1] = round(bnumber[0][0]*100,2)
    lst_avg[2] = round(bnumber[0][1]*100,2)
    lst_avg[3] = round(bnumber[0][2]*100,2)
    print("---saga---")
    print("Actual Types:",y_true)
    print("PredictTypes:",y_pred)
    print("AccuracyScore:",ac_score)
    print("Test Score:",lst_avg[0])
    print("HP1:",lst_avg[1])
    print("HP2:",lst_avg[2])
    print("HP3:",lst_avg[3])


    return lst_avg
```

```python

# name of the label (can be seen in the dataframe)
label = 'Respondents'

# list with feature names (V1, V2, V3, ...)
features = edata.columns.tolist()
features.remove(label)
features.remove('Survey_distribution')
features.remove('Buying possibility constant')


# change class labeling to 0 and 1
edata[label] = np.where( edata['Respondents'] > 35, 1, 0)
edata.head(5)
```

```python
from scipy.stats import pointbiserialr
from math import sqrt

def getMerit(subset, label):
    k = len(subset)

    # average feature-class correlation
    rcf_all = []
    for feature in subset:
        coeff = pointbiserialr( edata[label], edata[feature] )
        rcf_all.append( abs( coeff.correlation ) )
    rcf = np.mean( rcf_all )

    # average feature-feature correlation
    corr = edata[subset].corr()
    corr.values[np.tril_indices_from(corr.values)] = np.nan
    corr = abs(corr)
    rff = corr.unstack().mean()

    return (k * rcf) / sqrt(k + k * (k-1) * rff)
```

```python
best_value = -1
best_feature = ''
for feature in features:
    coeff = pointbiserialr( edata[label], edata[feature] )
    abs_coeff = abs( coeff.correlation )
    if abs_coeff > best_value:
        best_value = abs_coeff
        best_feature = feature

print("Feature %s with merit %.4f"%(best_feature, best_value))
```

```python
class PriorityQueue:
    def  __init__(self):
        self.queue = []

    def isEmpty(self):
        return len(self.queue) == 0

    def push(self, item, priority):
        """
        item already in priority queue with smaller priority:
        -> update its priority
        item already in priority queue with higher priority:
        -> do nothing
        if item not in priority queue:
        -> push it
        """
        for index, (i, p) in enumerate(self.queue):
            if (set(i) == set(item)):
                if (p >= priority):
                    break
                del self.queue[index]
                self.queue.append( (item, priority) )
                break
        else:
            self.queue.append( (item, priority) )

    def pop(self):
        # return item with highest priority and remove it from queue
        max_idx = 0
        for index, (i, p) in enumerate(self.queue):
            if (self.queue[max_idx][1] < p):
                max_idx = index
        (item, priority) = self.queue[max_idx]
        del self.queue[max_idx]
        return (item, priority)
```

```python
# initialize queue
queue = PriorityQueue()

# push first tuple (subset, merit)
queue.push([best_feature], best_value)
```

```python
# list for visited nodes
visited = []

# counter for backtracks
n_backtrack = 0

# limit of backtracks
max_backtrack = 5
```

```python
# repeat until queue is empty
# or the maximum number of backtracks is reached
while not queue.isEmpty():
    # get element of queue with highest merit
    subset, priority = queue.pop()

    # check whether the priority of this subset
    # is higher than the current best subset
    if (priority < best_value):
        n_backtrack += 1
    else:
        best_value = priority
        best_subset = subset

    # goal condition
    if (n_backtrack == max_backtrack):
        break

    # iterate through all features and look of one can
    # increase the merit
    for feature in features:
        temp_subset = subset + [feature]

        # check if this subset has already been evaluated
        for node in visited:
            if (set(node) == set(temp_subset)):
                break
        # if not, ...
        else:
            # ... mark it as visited
            visited.append( temp_subset )
            # ... compute merit
            merit = getMerit(temp_subset, label)
            # and push it to the queue
            queue.push(temp_subset, merit)
```

```python
from sklearn import linear_model
from sklearn.metrics import accuracy_score
import numpy

def predict_ppdemand_li(df1,XC,field1,field2,field3,field4,field5):
    #sdata = create_sumdata()
    lst_avg = [0,0,0,0,0]
    fields = [field1,field2,field3,field4,field5]
    nfields = []
    XC1 = []
    for i in range(5):
        if XC[i] != 0:
            nfields.append(fields[i])
            XC1.append(XC[i])

    print("Input Attributes:",nfields)
    print("Input Field:",XC1)
    X1 = numpy.array(df1[nfields])
    #print(numpy.array(sdata[[field1,field2,field3,field4,field5]]))
    #liblinear
    Y1 = numpy.array(df1['Htypes'])
    #regr1 = linear_model.LogisticRegression(solver='saga', random_state=0)
    regr1 = linear_model.LogisticRegression(solver='liblinear', random_state=0)
    regr1.fit(X1, Y1)
    y_pred = regr1.predict(numpy.array(df1[nfields]))
    y_true = numpy.array(df1['Htypes'])
    bnumber = regr1.predict_proba(numpy.array([XC1]))
    ascore = regr1.score(X1, Y1)
    ac_score = accuracy_score(y_true, y_pred)

    lst_avg[0] = round(ascore,3)
    lst_avg[1] = round(bnumber[0][0]*100,2)
    lst_avg[2] = round(bnumber[0][1]*100,2)
    lst_avg[3] = round(bnumber[0][2]*100,2)
    print("---Liblinear---")
    print("Actual Types:",y_true)
    print("PredictTypes:",y_pred)
    print("AccuracyScore:",ac_score)
    print("Test Score:",lst_avg[0])
    print("HP1:",lst_avg[1])
    print("HP2:",lst_avg[2])
    print("HP3:",lst_avg[3])


    return lst_avg

def predict_ppdemand_saga(df1,XC,field1,field2,field3,field4,field5):
    #sdata = create_sumdata()
    lst_avg = [0,0,0,0,0]
    fields = [field1,field2,field3,field4,field5]
    nfields = []
    XC1 = []
    for i in range(5):
        if XC[i] != 0:
            nfields.append(fields[i])
            XC1.append(XC[i])

    print("Input Attributes:",nfields)
    print("Input Field:",XC1)
    X1 = numpy.array(df1[nfields])
    #print(numpy.array(sdata[[field1,field2,field3,field4,field5]]))
    #liblinear
    Y1 = numpy.array(df1['Htypes'])
    #regr1 = linear_model.LogisticRegression(solver='saga', random_state=0)
    regr1 = linear_model.LogisticRegression(solver='saga', random_state=0)
    regr1.fit(X1, Y1)
    y_pred = regr1.predict(numpy.array(df1[nfields]))
    y_true = numpy.array(df1['Htypes'])
    bnumber = regr1.predict_proba(numpy.array([XC1]))
    ascore = regr1.score(X1, Y1)
    ac_score = accuracy_score(y_true, y_pred)

    lst_avg[0] = round(ascore,3)
    lst_avg[1] = round(bnumber[0][0]*100,2)
    lst_avg[2] = round(bnumber[0][1]*100,2)
    lst_avg[3] = round(bnumber[0][2]*100,2)
    print("---saga---")
    print("Actual Types:",y_true)
    print("PredictTypes:",y_pred)
    print("AccuracyScore:",ac_score)
    print("Test Score:",lst_avg[0])
    print("HP1:",lst_avg[1])
    print("HP2:",lst_avg[2])
    print("HP3:",lst_avg[3])


    return lst_avg
```
