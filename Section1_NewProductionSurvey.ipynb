# Section 1: Read Google Survey Form and Data Wrangling 

Read data from Google form responsing, clean and manipulate data to be appropriate for analysis


```python
#import data from Google sheet which came form Google survey form. 

from __future__ import print_function
import gspread
from oauth2client.service_account import ServiceAccountCredentials
import pandas as pd 
import numpy as np
import json

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option("display.max_colwidth", None)
pd.set_option('display.colheader_justify', 'center')
pd.set_option('display.precision', 3)
pd.options.mode.chained_assignment = None  # default='warn'

SCOPE =  ['https://www.googleapis.com/auth/spreadsheets',
          'https://www.googleapis.com/auth/drive']

SPREADSHEET = "New Headphone Production Survey (Responses)"

# Authenticate using the signed key
credentials = ServiceAccountCredentials.from_json_keyfile_name("customer-survey-feedback-f58deed6af2f.json", SCOPE)

gc = gspread.authorize(credentials)
print("The following sheets are available")
for sheet in gc.openall():
    print("{} - {}".format(sheet.title, sheet.id))
# Open up the workbook based on the spreadsheet name
workbook = gc.open(SPREADSHEET)
# Get the first sheet
sheet = workbook.sheet1
# Extract all data into a dataframe
data = pd.DataFrame(sheet.get_all_records())
# Do some minor cleanups on the data
# Rename the columns to make it easier to manipulate
# The data comes in through a dictionary so we can not assume order stays the
# same so must name each column

display(data)
```

```python
print(data.shape)
```

```python
#Changed all columns name 

data.columns.values[0] = "timestamp" 
data.columns.values[1] = "Score" 
data.columns.values[2] = "Name" 
data.columns.values[3] = "Email" 
data.columns.values[4] = "Gender" 
data.columns.values[5] = "Age" 
data.columns.values[6] = "City" 
data.columns.values[7] = "Status" 
data.columns.values[8] = "Education"
data.columns.values[9] = "Occupation"
data.columns.values[10] = "HpAnswer"
data.columns.values[11:16] = ["HpAscore_1","HpAscore_2","HpAscore_3","HpAscore_4","HpAscore_5"]
data.columns.values[16:21] = ["HpBscore_1","HpBscore_2","HpBscore_3","HpBscore_4","HpBscore_5"]
data.columns.values[21:26] = ["HpCscore_1","HpCscore_2","HpCscore_3","HpCscore_4","HpCscore_5"]
data.columns.values[26] = "NumberHps" 
data.columns.values[27] = "InfoHps" 
data.columns.values[28] = "ActivityHps" 
data.columns.values[29] = "TimeHps" 
data.columns.values[30] = "PriceHps" 
data.columns.values[31] = "PlaceHps" 
data.columns.values[32] = "ProblemHps" 
data.columns.values[33] = "FactorHps" 
data.columns.values[34] = "HealthHps" 
data.columns.values[35] = "InnovationHps" 
data.columns = data.columns.str.strip()

display(data.columns)
print(data.shape)
```

```python
# Removed non-English alphabets from data and fault data.

data['timestamp'] = pd.to_datetime(data['timestamp']).dt.date
data['Age']= data['Age'].str.encode('ascii', 'ignore').str.decode('ascii').str.replace(r'[/,"years"]',"", regex=True)
data['Gender']= data['Gender'].str.encode('ascii', 'ignore').str.decode('ascii').str.replace(r'[()]',"", regex=True)
data['Status']= data['Status'].str.encode('ascii', 'ignore').str.decode('ascii').str.replace(r'[()]',"", regex=True)
data['Education']= data['Education'].str.encode('ascii', 'ignore').str.decode('ascii').str.replace(r'[()]',"", regex=True)
data['Occupation']= data['Occupation'].str.encode('ascii', 'ignore').str.decode('ascii').str.replace(r'[()]',"", regex=True)
data['HpAnswer']= data['HpAnswer'].str.encode('ascii', 'ignore').str.decode('ascii').str.replace(r'[()]',"", regex=True)
data['NumberHps']= data['NumberHps'].str.encode('ascii', 'ignore').str.decode('ascii').str.replace(r'[/,"pieces"]',"", regex=True)
data['NumberHps']= data['NumberHps'].str.replace("mor than 5","more than 5", regex=True)
data['InfoHps']= data['InfoHps'].str.encode('ascii', 'ignore').str.decode('ascii').str.replace(r'[()]',"", regex=True)
data['ActivityHps']= data['ActivityHps'].str.encode('ascii', 'ignore').str.decode('ascii').str.replace(r'[()]',"", regex=True)
data['TimeHps']= data['TimeHps'].str.encode('ascii', 'ignore').str.decode('ascii').str.replace(r'[()]',"", regex=True)
data['PriceHps']= data['PriceHps'].str.encode('ascii', 'ignore').str.decode('ascii').str.replace(r'[(),/,"THB"]',"", regex=True)
data['PlaceHps']= data['PlaceHps'].str.encode('ascii', 'ignore').str.decode('ascii').str.replace(r'[()]',"", regex=True)
data['ProblemHps']= data['ProblemHps'].str.encode('ascii', 'ignore').str.decode('ascii').str.replace(r'[()]',"", regex=True)
data['FactorHps']= data['FactorHps'].str.encode('ascii', 'ignore').str.decode('ascii').str.replace(r'[()]',"", regex=True)
data['HealthHps']= data['HealthHps'].str.encode('ascii', 'ignore').str.decode('ascii').str.replace(r'[()]',"", regex=True)
data['InnovationHps']= data['InnovationHps'].str.encode('ascii', 'ignore').str.decode('ascii').str.replace(r'[()]',"", regex=True)
data.replace("", 0, inplace=True)
data.dropna(subset = ['Name'], inplace=True)
data.drop(data[data['Name']=="Test"].index, inplace=True)
data.drop(data[data['Name']==0].index, inplace=True)
data.drop(data[data['Gender']==0].index, inplace=True)
data.drop(data[data['Age']==0].index, inplace=True)
data.drop(data[data['Status']==0].index, inplace=True)
data.drop(data[data['Occupation']==0].index, inplace=True)
data.drop(data[data['Education']==0].index, inplace=True)
data.drop(["Email Address"], axis = 1, inplace = True)

dfPersonal = data[['Name','Email','Gender','Age','City','Status','Education','Occupation']]
dfHpAns = data[['timestamp','Name','HpAnswer']]
dfCbehavior = data[['Name','NumberHps','InfoHps','ActivityHps','TimeHps','PriceHps','PlaceHps','ProblemHps','FactorHps','HealthHps','InnovationHps']]

display(data)
```

```python
# Update headphone answer using short alphabet
dfHpAns['HpAnswer'].replace("Headphone v.1- Bone conduction headphone",'v1',inplace=True)
dfHpAns['HpAnswer'].replace("Headphone v.2- In-ear True Wireless headphone",'v2',inplace=True)
dfHpAns['HpAnswer'].replace("Headphone v.3- Over-Ear Headphone",'v3',inplace=True)

display(dfHpAns['HpAnswer'])
```

```python
# Create new columns to collect the score 

dfHpAns.loc[(data['Name']==dfHpAns['Name'])& (dfHpAns['HpAnswer']=="v1"), "HpTS1"] = data['HpAscore_1']
dfHpAns.loc[(data['Name']==dfHpAns['Name'])& (dfHpAns['HpAnswer']=="v1"), "HpTS2"] = data['HpAscore_2']
dfHpAns.loc[(data['Name']==dfHpAns['Name'])& (dfHpAns['HpAnswer']=="v1"), "HpTS3"] = data['HpAscore_3']
dfHpAns.loc[(data['Name']==dfHpAns['Name'])& (dfHpAns['HpAnswer']=="v1"), "HpTS4"] = data['HpAscore_4']
dfHpAns.loc[(data['Name']==dfHpAns['Name'])& (dfHpAns['HpAnswer']=="v1"), "HpTS5"] = data['HpAscore_5']

dfHpAns.loc[(data['Name']==dfHpAns['Name'])& (dfHpAns['HpAnswer']=="v2"), "HpTS1"] = data['HpBscore_1']
dfHpAns.loc[(data['Name']==dfHpAns['Name'])& (dfHpAns['HpAnswer']=="v2"), "HpTS2"] = data['HpBscore_2']
dfHpAns.loc[(data['Name']==dfHpAns['Name'])& (dfHpAns['HpAnswer']=="v2"), "HpTS3"] = data['HpBscore_3']
dfHpAns.loc[(data['Name']==dfHpAns['Name'])& (dfHpAns['HpAnswer']=="v2"), "HpTS4"] = data['HpBscore_4']
dfHpAns.loc[(data['Name']==dfHpAns['Name'])& (dfHpAns['HpAnswer']=="v2"), "HpTS5"] = data['HpBscore_5']

dfHpAns.loc[(data['Name']==dfHpAns['Name'])& (dfHpAns['HpAnswer']=="v3"), "HpTS1"] = data['HpCscore_1']
dfHpAns.loc[(data['Name']==dfHpAns['Name'])& (dfHpAns['HpAnswer']=="v3"), "HpTS2"] = data['HpCscore_2']
dfHpAns.loc[(data['Name']==dfHpAns['Name'])& (dfHpAns['HpAnswer']=="v3"), "HpTS3"] = data['HpCscore_3']
dfHpAns.loc[(data['Name']==dfHpAns['Name'])& (dfHpAns['HpAnswer']=="v3"), "HpTS4"] = data['HpCscore_4']
dfHpAns.loc[(data['Name']==dfHpAns['Name'])& (dfHpAns['HpAnswer']=="v3"), "HpTS5"] = data['HpCscore_5']

dfHpAns['HpTS1']=dfHpAns['HpTS1'].astype(int)
dfHpAns['HpTS2']=dfHpAns['HpTS2'].astype(int)
dfHpAns['HpTS3']=dfHpAns['HpTS3'].astype(int)
dfHpAns['HpTS4']=dfHpAns['HpTS4'].astype(int)
dfHpAns['HpTS5']=dfHpAns['HpTS5'].astype(int)

display(dfHpAns.head(10))
```
